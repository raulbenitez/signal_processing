Generate a simple periodic signal: 

```
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 500, endpoint=False)
plt.plot(t, signal.square(2 * np.pi * 5 * t))
plt.ylim(-2, 2)
```

Generate a pulse train of exponential functions:

```
from scipy.ndimage import convolve1d

# exponential decay profile: 

def x(t):
    return np.exp(- 0.5* t) * (t >= 0)
    
# pulse train function with unit impulses:
imp = np.zeros(100000)
imp[np.arange(0,len(imp),5000)] = 1

# exponential profile
t = np.linspace(-2, 20, 2000)
y = x(t)

# 1d convolution:
response = convolve1d(imp, weights=y)

# plot stuff
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(14,10))
#plt.plot(imp)
plt.plot(response)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
